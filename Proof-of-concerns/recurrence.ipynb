```python
ids_cliente_with_assunto = {i["id_cliente"]:[i['id_assunto'] for j in data["data"] if j['id_cliente'] == i["id_cliente"]] for i in data["data"]}
```

```python
def my_func(item, id_cliente):
    return item['id_cliente'] == id_cliente
```

```python
ids_cliente_with_assunto = {}

for i in data["data"]:

    id_cliente = i['id_cliente']

    if id_cliente not in ids_cliente_with_assunto:

        ids_cliente_with_assunto[id_cliente] = list(filter(lambda x: my_func(x, id_cliente), data['data']))

ids_cliente_with_assunto = {i:[j['id_assunto'] for j in ids_cliente_with_assunto[i]] for i in ids_cliente_with_assunto}
```

```python
ids_cliente_with_assunto
```

```python
only_tags = []
for i in data["data"]:
   if i['id_assunto'] not in only_tags: only_tags.append(i['id_assunto']) 
```

```python
only_tags
```

```python
ids_cliente_with_assunto
```

```python
tag_mapping_for_id = {

}

for i in ids_cliente_with_assunto:

    tag_mapping_for_id[i] = []

    for j in only_tags:

        if  ids_cliente_with_assunto[i].count(j) >0:

    
            tag_mapping_for_id[i].append({j:ids_cliente_with_assunto[i].count(j)})
```

```python
tag_mapping_for_id
```

```python
with open('ids_com_qtd_por_assunto.json','w') as file:
    types = json.dumps(tag_mapping_for_id)
    file.write(types)
```
